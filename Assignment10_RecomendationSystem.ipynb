```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

# EDA

```python
book=pd.read_csv("/content/sample_data/book.csv",encoding='latin-1')
book.head()
```

```python
book_new=book.drop('Unnamed: 0',axis=1)
book_new=book_new.rename({'User.ID':'User_id','Book.Title':'Book_Title','Book.Rating':'Book_Rating'},axis=1)
```

```python
book_new.head()
```

```python
book_new.info()
```

```python
print("list of uniqe user id:  ",book_new.User_id.unique())
print("total number of user:   ",book_new.User_id.nunique())
```

```python
book_new=book_new.sort_values('User_id')
book_new.head()
```

```python
book_new.Book_Rating.value_counts()
```

```python
len(book_new.Book_Title.unique())
```

```python
book_df=book_new.pivot_table(index="User_id",
                       columns="Book_Title",
                       values="Book_Rating").reset_index(drop=True)
book_df
```

```python
book_df.fillna(0,inplace=True)
book_df
```

```python
#calculating cosine similarity between users
reader_sim=1-pairwise_distances(book_df.values,metric='cosine')
reader_sim
```

```python
reader_sim_df=pd.DataFrame(reader_sim)
reader_sim_df
```

```python
reader_sim_df.index=book_new.User_id.unique()
reader_sim_df.columns=book_new.User_id.unique()
reader_sim_df
```

```python
np.fill_diagonal(reader_sim,0)
reader_sim_df.iloc[0:5, 0:5]
```
